```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ThousandAI/pycs4001/main/class09/k-means.csv")
data
```

```python
x1 = np.array(data.iloc[:,0])
x2 = data.iloc[:,1].values
print(type(x1))
print(type(x2))
```

```python
plt.scatter(x1, x2, s=2)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, max_iter=1000, random_state=1)
data = np.array(data)
np.random.shuffle(data)
kmeans.fit(data)
y_kmeans = kmeans.predict(data)
print(y_kmeans)

```

```python
plt.scatter(data[y_kmeans==0][:,0], data[y_kmeans==0][:,1], s = 2, color="red")
plt.scatter(data[y_kmeans==1][:,0], data[y_kmeans==1][:,1], s = 2, color="aqua")
plt.scatter(data[y_kmeans==2][:,0], data[y_kmeans==2][:,1], s = 2, color="gold")
plt.show()
```
