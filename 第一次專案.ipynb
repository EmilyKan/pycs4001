```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
# 資料來源: https://www.kaggle.com/datasets/victorsoeiro/netflix-tv-shows-and-movies?select=credits.csv
df = pd.read_csv("/content/netflix_titles.csv")


label = 'Movies','TV Shows'
color = ['#b2d8f7','#d8baf7']
explode = (0.1,0)
sizes = df.type.value_counts()
plt.pie(sizes,explode = explode, colors= color, labels =label,shadow = True,autopct ='%1.2f%%')
plt.axis('equal');

df_movies = df[df['type']== 'Movie'].reset_index() 
plt.figure(figsize=(10,6))
top_countries = df_movies['country'].value_counts().rename_axis('country').reset_index(name = 'counts')[:5]
sns.barplot(data = top_countries, x= 'counts',y = 'country')
plt.title('Top 5 country on Netflix');

```
